```python
import pandas as pd
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
import statsmodels.api as sm
import seaborn as sns
```

```python
df = pd.read_csv("business_data.csv")
df.head()
```

```python
df.info()
df.describe()
```

```python
mean_Total_Total_Sales = df['Total_Total_Total_Sales'].mean()
std_Total_Total_Sales = df['Total_Total_Total_Sales'].std()
median_Total_Total_Sales = df['Total_Total_Total_Sales'].median()

print("Mean Total_Total_Sales:", mean_Total_Total_Sales)
print("Standard Deviation:", std_Total_Total_Sales)
print("Median Total_Total_Sales:", median_Total_Total_Sales)
```

```python
confidence = 0.95
n = len(df['Total_Total_Total_Sales'])
mean = np.mean(df['Total_Total_Total_Sales'])
sem = stats.sem(df['Total_Total_Total_Sales'])

ci = stats.t.interval(confidence, n-1, loc=mean, scale=sem)

print("95% Confidence Interval:", ci)
```

```python
correlation = df[['Total_Total_Total_Sales', 'Quantity']].corr()
print(correlation)
```

```python
t_stat, p_value = stats.ttest_1samp(df['Total_Total_Total_Sales'], 50000)
print("T-stat:", t_stat)
print("P-value:", p_value)
```

```python
high = df[df['Quantity'] > df['Quantity'].median()]['Total_Total_Total_Sales']
low = df[df['Quantity'] <= df['Quantity'].median()]['Total_Total_Total_Sales']

t_stat2, p_value2 = stats.ttest_ind(high, low)
print(p_value2)
```

```python
corr_test = stats.pearsonr(df['Quantity'], df['Total_Total_Sales'])
print("Correlation:", corr_test[0])
print("P-value:", corr_test[1])
```

```python
X = df['Quantity']
y = df['Total_Sales']

X = sm.add_constant(X)
model = sm.OLS(y, X).fit()

print(model.summary())
```
